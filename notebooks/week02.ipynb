```python
import numpy as np

# 坐标
x = np.array([-1, 0, 2, 3, 4])
y = np.array([0, 1, 0, 1, 2])
# 构建范德蒙矩阵
R = np.vander(x, 5)
# 解方程
a = np.linalg.solve(R, y)
# 系数结果
print(a)
```

```python
import matplotlib.pyplot as plt
import numpy as np

# 线性方程的系数
x = np.array([[-1, 1, -1, 1], [3, -2, 1, 0], [3, 2, 1, 0], [1, 1, 1, 1]])
# 右边的y
y = np.array([-2, 1, -1, 0])
# 求解
r = np.linalg.solve(x, y)
a, b, c, d = r
print(a, b, c, d)


def p(x):
    return a * x**3 + b * x**2 + c * x + d


# 在【-1，1】之间生存1000个等距的点
x0 = np.linspace(-1, 1, 1000)
y0 = p(x0)
# 图像
plt.plot(x0, y0)
plt.xlabel("x0")
plt.ylabel("y0")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# 年份
x = np.array([1990, 2000, 2010, 2020])
# 各个国家的人口数据
USA = np.array([252.120, 281.711, 309.011, 331.003])
India = np.array([873.278, 1056.576, 1234.281, 1380.004])
Poland = np.array([37.960, 38.557, 38.330, 37.847])
# 估计美国2005年人口
func = np.polyfit(x, USA, 3)
# 预测美国人口的函数
Americans = np.poly1d(func)
USA2005 = Americans(2005)
print(f"2005年美国人口{USA2005:.3f}百万")
# 计算波兰人口在这30年的峰值
func1 = np.polyfit(x, Poland, 3)
Polands = np.poly1d(func1)
# 求一阶导
daohanshu = Polands.deriv()
# 一阶导数为0的点
roots = np.roots(daohanshu)
# 选出实数根在1990-2020之间的
realroots = [i for i in roots if np.isreal(i) and 1990 <= i.real <= 2020]
# 求二阶导判断是否为极大值点
erjiedaohanshu = daohanshu.deriv()
maxrealroots = []
for i in realroots:
    x4 = erjiedaohanshu(i.real)
    if x4 < 0:
        maxrealroots.append(i.real)
years = [i for i in maxrealroots]
for year in years:
    print(f"波兰峰值年份：{year:.0f}")
# 印度
func2 = np.polyfit(x, India, 3)
Indian = np.poly1d(func2)
# 生成坐标
x0 = np.linspace(1990, 2020, 300)
y0 = Indian(x0)
plt.scatter(x, India, label="yuanshi shuju")
plt.plot(x0, y0, label="shangci chazhi quxian")
plt.xlabel("year")
plt.ylabel("populaiton")
plt.legend()
plt.show()
```
