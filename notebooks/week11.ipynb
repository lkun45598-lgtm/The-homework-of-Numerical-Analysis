```python
import numpy as np
from numpy.linalg import matrix_rank
from scipy.linalg import eigvals

A = np.array([[2, -1, 0], [-1, 2, -1], [0, -1, 2]])

B = np.array([[2, -1, -1], [-2, 2, -1], [-1, -2, 2]])

C = np.array([[2, -1, -1], [-1, 2, -1], [-1, -1, 2]])

D = np.array([[3, 1, 0, 0], [1, 3, 1, 0], [0, 1, 3, 1], [0, 0, 1, 3]])

E = np.array([[4, -3, -2, -1], [-2, 4, -2, -1], [-1, -2, 4, -1], [-1, -2, -1, 4]])


def fun(matrix, k):
    lamadas = eigvals(matrix)
    for i in lamadas:
        lamada_matrix = i * np.eye(matrix.shape[0]) - matrix
        rank = matrix_rank(lamada_matrix)
        if rank < matrix.shape[0]:
            print(f":{i}{k}")
        else:
            print(f":{i}{k}")
    print()


fun(A, "A")
fun(B, "B")
fun(C, "C")
fun(D, "D")
fun(E, "E")
```

```python
import numpy as np

lamada = np.arange(-9.6, 10.5, 1.0)


n = len(lamada)
np.random.seed(42)
random_matrix = np.random.randn(n, n)
Q, R = np.linalg.qr(random_matrix)
D = np.diag(lamada)
A = Q @ D @ Q.T

# 200
for _ in range(200):
    Q, R = np.linalg.qr(A)
    A = R @ Q

final_lamada = np.diag(A)
final_abs = np.abs(final_lamada)

sorted_abs = np.sort(final_abs)[::-1]

print([float(f"{x:.4f}") for x in final_lamada])
print([float(f"{x:.4f}") for x in final_abs])
```

```python
import matplotlib.pyplot as plt
import numpy as np

A = np.random.randn(60, 60)
lamada = np.linalg.eigvals(A)
plt.scatter(np.real(lamada), np.imag(lamada), c="r", marker="D", label="lamada")
plt.xlabel("real number")
plt.ylabel("imaginary number")
plt.title("the lamada of matrix A")
plt.axis("equal")
plt.tight_layout()
plt.legend()
plt.show()
```

```python
n = 100
for _ in range(n):
    E = np.random.randn(60, 60)
    disturbed_matrix = A + 0.05 * E
    lamada_A = np.linalg.eigvals(disturbed_matrix)
    plt.scatter(np.real(lamada_A), np.imag(lamada_A), c="b", marker="o", alpha=0.2)
plt.scatter(np.real(lamada), np.imag(lamada), c="r", marker="D", label="lambda of A")
plt.axis("equal")
plt.title("the disturbed number add in the matrix")
plt.xlabel("Real number")
plt.ylabel("Imaginary number")
plt.legend()
plt.tight_layout()
plt.show()
```

```python
T = np.triu(A)
lamada_T = np.linalg.eigvals(T)
plt.scatter(
    np.real(lamada_T), np.imag(lamada_T), c="g", marker="s", label="lamada of T"
)
plt.xlabel("real number")
plt.ylabel("imaginary number")
plt.title("the lamada of T")
plt.axis("equal")
plt.tight_layout()
plt.legend()
plt.show()
```

```python
n = 100
for _ in range(n):
    E = np.random.randn(60, 60)
    disturbed_matrix = T + 0.05 * E
    lamada_A = np.linalg.eigvals(disturbed_matrix)
    plt.scatter(np.real(lamada_A), np.imag(lamada_A), c="b", marker="o", alpha=0.2)
plt.scatter(
    np.real(lamada_T), np.imag(lamada_T), c="r", marker="D", label="lambda of T"
)
plt.axis("equal")
plt.title("the disturbed number add in the matrix")
plt.xlabel("Real number")
plt.ylabel("Imaginary number")
plt.legend()
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 1, 1000)


def matrix(n):
    matrix_x = np.zeros((len(x), n))
    for j in range(n):
        matrix_x[:, j] = x**j
    return matrix_x


A1 = matrix(1)
A2 = matrix(2)
A3 = matrix(3)

svd_A1 = np.linalg.svd(A1, compute_uv=False)
svd_A2 = np.linalg.svd(A2, compute_uv=False)
svd_A3 = np.linalg.svd(A3, compute_uv=False)

print(f"A1:{svd_A1}")
print(f"A1:{svd_A2}")
print(f"A1:{svd_A3}")
```

```python
A40 = matrix(40)
svd_40 = np.linalg.svd(A40, compute_uv=False)
# 
n = [int(i) for i in range(40)]
plt.semilogy(n, svd_40, "o-", label="svd of A40")
plt.xlabel("n")
plt.ylabel("svd")
plt.title("the svd of A40")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.linalg import eigvals

A = np.array([[1, 0, -2], [0, 2, 0], [-2, 0, 1]])

# 
random_matrix = np.random.randn(500, A.shape[0])

zhiyu = []
for x in random_matrix:
    a = np.vdot(x, A @ x)
    b = np.vdot(x, x)

    # 
    c = a / b
    zhiyu.append(c)

values = np.array(zhiyu)

# 
plt.scatter(values.real, values.imag, alpha=0.8, label="zhiyu")
plt.title("Values of Matrix A")
plt.xlabel("Real number")
plt.ylabel("Imaginary number")
plt.legend()
plt.grid(True)
plt.show()

value = np.linalg.eigvals(A)
print(f"{value}")
```

```python
import numpy as np

A = np.array([[1, 5], [5, 1]])

U, S, V = np.linalg.svd(A)

# 
s_max = S[0]
u1 = U[:, 0].reshape(-1, 1)
v1 = V[0, :].reshape(1, -1)

#  1 
matrix = s_max * np.dot(u1, v1)

print(matrix)
```
