```python
import numpy as np 
import matplotlib.pyplot as plt
from scipy.optimize import root
import sympy as sp
#绘图
def f(x):
    return x**2 - np.exp(-x)
x = np.linspace(-2,2,100)
y = f(x)
plt.plot(x,y,label='x**2-exp(-x)')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()
root1 = root(f,1)
root1 = root1.x[0]
print(f"根:{root1}")
#条件数
x0 = sp.symbols('x')
f1 = x0**2-sp.exp(-x0)
df = sp.diff(f1,x0)
cond = np.abs(root1*df.subs(x0,root1)/(f(root1)+1e-10))
print(f"条件数:{cond}")
```

```python
import numpy as np
from scipy.optimize import root
# 定义年金终值的函数
def func(P, r, n):
    return (12 * P / r) * ((1 + r / 12) ** n - 1)
n = 300
target_money = 1000000
# 不同的贡献值
P = np.arange(500, 1050, 50)
# 存储结果
profits = []
for p in P:
    # 构造求解的方程
    def func1(r):
        return func(p, r, n) - target_money
    # 初始猜测值
    guess = 0.05
    # 使用root函数求解
    sol = root(func1,guess)
    profits.append(sol.x[0])

# 打印结果表格
print("贡献值: P  年利率: r")
for P, r in zip(P,profits):
    print(f"{P}        {r}")
```

```python
import numpy as np
import matplotlib.pyplot as plt
def g(x):
    return (x*x+3.5)/4
# 定义定点迭代函数
def deidai_function(x0,max_times=100,tolerance=1e-6):
    xn = x0
    for i in range(max_times): 
        xn_next = g(xn)
        if abs(xn_next - xn) < tolerance:
            return xn_next
        xn = xn_next
    print("迭代未在指定次数内收敛")
    return None

# 尝试不同的初始值
initial_values = [0, 10]
for x0 in initial_values:
    result = deidai_function(x0)
    if result is not None:
        print(f"初始值为 {x0} 时收敛到:{result}")
```

```python
import numpy as np
import matplotlib.pyplot as plt
def g(x):
    return 2*x-3*x**2
# 初始值
x0 = 0.4
# 固定点
p = 1/3
#迭代次数
n = 10
x = [x0]
#误差
error = [abs(x0-p)]
for _ in range(n):
    x_next = g(x[-1])
    error_0 = abs(x_next-p)
    x.append(x_next)
    error.append(error_0)
x1 = np.array([i for i in np.arange(11)])
#绘图
plt.semilogy(x1,error)
plt.xlabel('n')
plt.ylabel('wucai')
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import root
import sympy as sp
# 绘图
def funx(x):
    return x*x-np.exp(-x)
x = np.linspace(-2,2,100)
y = funx(x)
plt.plot(x,y,label='x*x-exp(-x)')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()
# 定义符号变量和函数
x0 = sp.symbols('x')
f = x0 * x0 - sp.exp(-x0)
# 定义导数函数
def dfun(f, x0):
    f_n = f
    f_n = sp.diff(f_n,x0)
    return sp.lambdify(x0,f_n)  
# 定义牛顿法函数
def fun(fun1,dfun1,x,n,ftol=100*np.finfo(float).eps,xtol=100 * np.finfo(float).eps):
    x0 = [float(x)]
    y = fun1(x0[0])
    delta_x = float("inf")
    k = 0
    while (abs(delta_x) > xtol) and (abs(y) > ftol):
        #计算导数值
        df = dfun1(x0[k])  
        delta_x = -y / df
        #更新x0
        x0.append(x0[k] + delta_x)  
        k += 1
        y = fun1(x0[k])
        if k == n:
            break
    return x0
# 计算导数函数
dfun1 = dfun(f, x0)
# 使用牛顿法求解
x1 = fun(funx, dfun1, 0.5, 40)
for i in x1:
    print(i)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp
def fun(x):
    return 2*x-np.tan(x)
x = np.linspace(-0.2,1.4)
y = fun(x)
plt.plot(x,y,label='2*x-tan(x)')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()
x0 = sp.Symbol('x')
f = x0*2-sp.tan(x0)
def df(f,x0):
    f1 = f
    f1 = sp.diff(f1,x0)
    return sp.lambdify(x0,f1)
#定义牛顿函数
def nwetwo(f,df,n,x,ftol = 100*np.finfo(float).eps,xtol = 100*np.finfo(float).eps):
    x1 = [float(x)]
    y = f(x1[0])
    k = 0
    delta = float("inf")
    while (abs(delta)>xtol) and (abs(y)>ftol):
        delta = -y/df(x1[k])
        x2 = x1[k]+delta
        x1.append(x2)
        k+=1
        y = f(x1[k])
        if k==n:
            break
    return x1
dfun = df(f,x0)
x1 = nwetwo(fun,dfun,40,.4)
for i in x1:
    print(i)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp
def fun(x):
    return np.exp(x+1)-x-2
x = np.linspace(-2,2,100)
y = fun(x)
plt.plot(x,y,label='exp(x+1)-x-2')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.legend()
plt.show()
x0 = sp.symbols('x')
f = sp.exp(x0+1)-x0-2
def df(f,x0):
    f1 = sp.diff(f,x0)
    return sp.lambdify(x0,f1)
def newton(fun,dfun,x,n,xtol = np.finfo(float).eps,ftol = np.finfo(float).eps):
    x0 = [float(x)]
    y = fun(x0[0])
    k = 0
    delta = float('inf')
    while abs(delta)>xtol and abs(y)>ftol:
        delta = -y/dfun(x0[k])
        x = x0[k]+delta
        x0.append(x)
        k+=1
        y = fun(x0[k])
        if k==n:
            break
    return x0
dfun = df(f,x0)
x1 = newton(fun,dfun,-1.2,40)
for i in x1:
    print(i)
```

```python
import numpy as np
import matplotlib.pyplot as plt
# 定义函数
def fun(x):
    return x * x - np.exp(-x)
# 绘制函数图像
x = np.linspace(-2,2,100)
y = fun(x)
plt.plot(x, y, label="f(x) = x^2 - exp(-x)")
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.legend()
plt.show()
# 用割线方法求根
def secant(f, x1, x2, n=40, ftol=100 * np.finfo(float).eps, xtol=100 * np.finfo(float).eps):
    x = [float(x1), float(x2)]
    y1 = f(x1)
    delta_x, y2 = float('inf'), float('inf')
    k = 2
    while (abs(delta_x) > xtol) and (abs(y2) > ftol):
        y2 = f(x[k - 1])
        delta_x = -y2 * (x[k - 1] - x[k - 2]) / (y2 - y1)
        x.append(x[k - 1] + delta_x)
        k += 1
        y1 = y2
        if k == n:
            break

    return x
#调用
root_estimates = secant(fun, 0, 0.5)
for i in root_estimates:
    print(i)
```
