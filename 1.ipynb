```python
x=[float(i) for i in input().split()]
def average(x):
    n=len(x)
    avex=sum(x)/n
    s2=0
    for i in range(n):
        s2+=(x[i]-avex)**2
    s2=s2/(n-1)
    print(s2,avex)
average(x)
```

```python
x=[float(i) for i in input().split()]
y=[float(i) for i in input().split()]
def polygonarea(x,y):
    a=len(x)
    A=0
    for i in range(a-1):
        A+=0.5*abs(x[i]*y[i+1]-x[i+1]*y[i])
    print(A)
polygonarea(x,y)
```
