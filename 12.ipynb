```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.sparse import spdiags,csr_matrix

data_A = [np.ones(50),-2 * np.ones(50),np.ones(50)]
pianyi = [-1,0,1]
A = spdiags(data_A,pianyi,50,50).tocsr()

data_B = [np.ones(50),-2 * np.ones(50),np.ones(50)]
pianyi = [-1,0,1]
B = spdiags(data_B,pianyi,50,50).tocsr()
B[0,49] = 1
B[49,0] = 1


plt.spy(A,markersize = 4,label = 'A')
plt.xlabel('column')
plt.ylabel('row')
plt.title('matrix B')
plt.legend()
plt.show()

plt.spy(B,markersize = 3,label = 'B')
plt.xlabel('column')
plt.ylabel('row')
plt.title('matrix B')
plt.legend()
plt.show()
```

```python
import numpy as np

 

def poweriter(A, numiter):

    n = A.shape[0]
    x = np.random.randn(n)
    x = x / np.linalg.norm(x, np.inf)  # 归一化
    β = np.zeros(numiter)
    
    for k in range(numiter):
        y = A @ x  
        m = np.argmax(np.abs(y))
        β[k] = y[m] / x[m]
        x = y / y[m]  # 归一化，使无穷范数为1
    
    return β, x

A = np.array([[1.1,1],
              [0.1,2.4]])

B = np.array([[2,1],
              [1,0]])

C = np.array([[6,5,4],
              [5,4,3],
              [4,3,2]])

D = np.array([[8,-14,0,-14],
              [-8,1,1,1],
              [-4,-2,0,2],
              [8,-7,-1,-7]])

B1,X1 = poweriter(A,20)
B2,X2 = poweriter(B,20)
B3,X3 = poweriter(C,20)
B4,X4 = poweriter(D,20)

print(f"B1:{B1}\n\nX1:{X1}\n\n")
print(f"B2:{B2}\n\nX2:{X2}\n\n")
print(f"B3:{B3}\n\nX3:{X3}\n\n")
print(f"B4:{B4}\n\nX4:{X4}\n\n")
```
